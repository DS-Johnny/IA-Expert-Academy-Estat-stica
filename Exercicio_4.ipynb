# Exercício 4

O objetivo deste exercício é calcular o coeficiente e a taxa de evasão utilizando a tabela abaixo. Faça os cálculos manuais e depois implemente em Python

```python
import pandas as pd
```

```python
dados = {
    'Ano graduação' : ['1º', '2º', '3º', '4º'],
    'Matrículas março' : [70, 50, 47, 23],
    'Matrículas novembro' : [65, 48, 40, 22]
}

dataset = pd.DataFrame(dados)
dataset
```

```python
dataset['Coeficiente Evasão'] = (dataset['Matrículas março'] - dataset['Matrículas novembro']) / dataset['Matrículas março']
```

```python
dataset['Taxa evasão'] = dataset['Coeficiente Evasão'] * 100
```

```python
dataset
```

```python
for _, i in dataset.iterrows():
    print(f'A taxa de evasão do {i['Ano graduação']} ano foi de {i['Taxa evasão']:.2f}%')

print(f'\nO taxa de evasão da escola foi de {((dataset['Matrículas março'].sum() - dataset['Matrículas novembro'].sum()) / dataset['Matrículas março'].sum()) *100:.2f}%')
```
